2\. Find the outer product of the following vectors:

```python
u = np.array([1, 3, 5, 7])
v = np.array([2, 4, 6, 8])
```

Use different methods to do this:

   1. Using the function `outer` in numpy
   2. Using a nested `for` loop or a list comprehension
   3. Using numpy broadcasting operations

````python
u = np.array([1, 3, 5, 7])
v = np.array([2, 4, 6, 8])
#1 Using the function outer in numpy
pro1=np.outer(u,v)
print(pro1)
#2 Using a nested for loop
pro2=pro1*0
for i in range(0,4):
    for j in range(0,4):
           pro2[i,j]=u[i]*v[j]
print(pro2)
#3 Using numpy broadcasting operations
pro3=u[:, np.newaxis] * v
print(pro3)
````

7\. Prime numbers sieve: compute the prime numbers in the 0-N (start with N=99) range with a sieve (mask).
  * Check the performances (timeit); how does it scale with N?
  * Implement the optimization suggested in the [sieve of Eratosthenes](https://en.wikipedia.org/wiki/Sieve_of_Eratosthenes)

````python

def primeN(N):
    prime=[]
    prime.append(2)
    for num in range(3,N):
        ind=0
        for i in range(2,num):
            if num%i==0:
                ind+=1
        if ind==0:
            prime.append(num)
    return prime

def opt(N):
    mask=[True for i in range(1,N)]
    mask[0]=False
    mask[1]=False
    i=2
    while i*2<N:
        if mask[i]:
            for l in range(i,N-1,i):
                mask[l]=False
            mask[i]=True
        i+=1
    return [n for n in range(N-1)if mask[n]]


N=99

print(primeN(N))
print(opt(N))
print("non optimized")
%timeit primeN(N)
%timeit primeN(N*10)
%timeit primeN(N*50)
%timeit primeN(N*100)
print("optimized")
%timeit opt(N)
%timeit opt(N*10)
%timeit opt(N*50)
%timeit opt(N*100)

````
